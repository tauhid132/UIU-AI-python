```python
from numpy import genfromtxt
data_path = '/content/gdrive/MyDrive/AI_LAB/Assignment-1/iris.csv'
my_data = genfromtxt(data_path, delimiter=',')
```

```python
my_data
```

```python
from random import seed
import random

seed(1)

train_set = []
val_set = []
test_set = []
random.shuffle(my_data)
for S in my_data:
  value = random.random()
  if value>=0 and value<=0.7:
    train_set.append(S)
  elif value>=0.7 and value<=0.85:
    val_set.append(S)
  else:
    test_set.append(S)

```

```python
from math import sqrt
def maximum(a, b, c):
  
    if (a >= b) and (a >= c):
        largest = 0
  
    elif (b >= a) and (b >= c):
        largest = 1
    else:
        largest = 2
          
    return largest
#def eu_dis(A, B):
#  return sqrt((A[0]-B[0])**2+(A[1]-B[1])**2+(A[2]-B[2])**2+(A[3]-B[3])**2)
```

```python
import numpy as np

K = 5
correct = 0
for V in val_set:
  L = []
  for T in train_set:
    source = np.array(V[0:4])
    destination = np.array(T[0:4])
    dist = np.linalg.norm(source - destination)
    L.append([T, dist])
  L.sort(key=lambda L:L[1])
  classZero = 0
  classOne = 0
  classTwo = 0
  for i in range(K):
    temp_list = []
    temp_list = L[i][0]
    if temp_list[4]==0:
      classZero+=1
    elif temp_list[4]==1:
      classOne+=1
    elif temp_list[4]==2:
      classTwo+=1
  predictionClass = maximum(classZero, classOne, classTwo)
  #print(L)
  #print(predictionClass , V[4])
  if predictionClass == V[4]:
    correct+=1
    #print("Correct")
  #else:
    #print("Incorrect")
accuracy = correct / len(val_set) * 100 
print("Validation Accuracy:")
print(accuracy)
  
   
    
```

```python
K = 5
L = []
test_correct = 0
for V in test_set:
  L = []
  for T in train_set:
    source = np.array(V[0:4])
    destination = np.array(T[0:4])
    dist = np.linalg.norm(source - destination)
    L.append([T, dist])
  L.sort(key=lambda L:L[1])
  classZero = 0
  classOne = 0
  classTwo = 0
  for i in range(K):
    temp_list = []
    temp_list = L[i][0]
    if temp_list[4]==0:
      classZero+=1
    elif temp_list[4]==1:
      classOne+=1
    elif temp_list[4]==2:
      classTwo+=1
  predictionClass = maximum(classZero, classOne, classTwo)
  
  if predictionClass == V[4]:
    test_correct+=1
    
test_accuracy = test_correct / len(test_set) * 100 
print("Test Accuracy:")
print(test_accuracy)
```
