```python
from numpy import genfromtxt
data_path = '/content/gdrive/MyDrive/AI_LAB/Assignment-1/diabetes.csv'
my_data = genfromtxt(data_path, delimiter=',')
```

```python
from random import seed
import random

seed(1)

train_set = []
val_set = []
test_set = []
random.shuffle(my_data)
for S in my_data:
  value = random.random()
  if value>=0 and value<=0.7:
    train_set.append(S)
  elif value>=0.7 and value<=0.85:
    val_set.append(S)
  else:
    test_set.append(S)
```

```python
import numpy as np
K = 1
L = []
error = 0
for V in val_set:
  L = []
  for T in train_set:
    source = np.array(V[0:10])
    destination = np.array(T[0:10])
    dist = np.linalg.norm(source - destination)
    L.append([T, dist])
  L.sort(key=lambda L:L[1])
  total = 0
  for i in range(K):
    temp_list = []
    temp_list = L[i][0]
    total = total + temp_list[10]
  avg_result = total / K
  error = error + (V[10]-avg_result)**2
  #print(V[10], avg_result)
mean_sq_error = error/len(val_set)
print("Validation Mean squared Error")
print(mean_sq_error)
```

```python
import numpy as np
K = 1
error = 0
for V in test_set:
  L = []
  for T in train_set:
    source = np.array(V[0:10])
    destination = np.array(T[0:10])
    dist = np.linalg.norm(source - destination)
    L.append([T, dist])
  L.sort(key=lambda L:L[1])
  total = 0
  for i in range(K):
    temp_list = []
    temp_list = L[i][0]
    total = total + temp_list[10]
  avg_result = total / K
  error = error + (V[10]-avg_result)**2
  #print(V[10], avg_result)
mean_sq_error = error/len(test_set)
print("Test Mean squared Error")
print(mean_sq_error)
```
